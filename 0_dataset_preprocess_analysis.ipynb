## Preprocessing

### Joining all files into a single one

```python
def join_dfs(folder_path):
    dataframes = []
    
    #Iterate over the files
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):  
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_csv(file_path)
            dataframes.append(df)
            
    #Combine all dataframes
    combined_df = pd.concat(dataframes, ignore_index=True)
    return combined_df
```

Join Questions

```python
folder_path_questions = 'stackoverflow_2019/questions'
df_questions = join_dfs(folder_path_questions)
```

Join Answers

```python
folder_path_answer = 'stackoverflow_2019/answers'
df_answers = join_dfs(folder_path_answer)
```

Clean tags

```python
# Replace NaN with empty strings (if applicable) before processing
df_questions["QuestionTags"] = df_questions["QuestionTags"].fillna("")

# Transform the tags into a list format
df_questions["QuestionTags"] = df_questions["QuestionTags"].apply(lambda tags: tags.split("><") if tags else [])
df_questions["QuestionTags"] = df_questions["QuestionTags"].apply(lambda tags: [tag.strip("<>") for tag in tags])
```

Save DFs

```python
df_questions.to_csv("dataset/questions_2019.csv", index=False)
df_answers.to_csv("dataset/answers_2019.csv", index=False)
```

```python
# Ordering dataframe before split
# Convert 'QuestionCreationDate' to datetime format using .loc
df_question_filtered.loc[:, 'QuestionCreationDate'] = pd.to_datetime(df_question_filtered['QuestionCreationDate'])

# Sort the DataFrame by 'QuestionCreationDate' without using inplace
df_question_filtered = df_question_filtered.sort_values(by='QuestionCreationDate')
```

```python
# Determine the split index
split_index = int(len(df_question_filtered) * 0.9) #10% for test 

# Split the DataFrame into training and testing sets
train_data = df_question_filtered.iloc[:split_index]
test_data = df_question_filtered.iloc[split_index:]
```

```python
all_tags = [tag for tags in train_data["QuestionTags"] for tag in tags]
tag_counts = Counter(all_tags)
tag_counts_df_train = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values(by="Count", ascending=False).head(10)
tag_counts_df_train["type"] = "train"
total_train = tag_counts_df_train['Count'].sum()
tag_counts_df_train['Percentage'] = (tag_counts_df_train['Count'] / total_train) * 100


all_tags = [tag for tags in test_data["QuestionTags"] for tag in tags]
tag_counts = Counter(all_tags)
tag_counts_df_test = pd.DataFrame(tag_counts.items(), columns=["Tag", "Count"]).sort_values(by="Count", ascending=False).head(10)
tag_counts_df_test["type"] = "test"
total_test = tag_counts_df_test['Count'].sum()
tag_counts_df_test['Percentage'] = (tag_counts_df_test['Count'] / total_test) * 100



tag_counts_df = pd.concat([tag_counts_df_train, tag_counts_df_test])

```

```python
# Sorting the data by 'Percentage' in descending order
combined_data_sorted = tag_counts_df.sort_values(by='Percentage', ascending=False)

# Plot side-by-side bar chart with the same styling
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=combined_data_sorted, x='Tag', y='Percentage', hue='type', palette=['#6495ED', '#FFA07A'], hue_order=['train', 'test'])

# Add title and axis labels
plt.title('Train vs Test Tag Distribution', fontsize=24, fontweight="bold", pad=20)
plt.xlabel('Tags', fontsize=20)
plt.ylabel('Percentage of the top 5 Tags', fontsize=20)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=18)
plt.yticks(fontsize=18)
# Format y-axis ticks as integers with percentage sign
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{int(y)}%'))

# Customize legend
plt.legend(title='Dataset Type', fontsize=18, title_fontsize=20)

# Remove top and right borders
sns.despine()

# Show the plot
plt.show()
```

```python

# Define a new custom palette with distinct colors for all types
custom_palette = {
    "top_users": "black", 
    "collaborative": "#E15759",  # Strong Red
    "content": "#F28E2B",  # Warm Orange
    "nlp_sentence_transformer_user_profile": "#86BBD8",  # Light Blue
    "nlp_sentence_transformer_question": "#5AA469",  # Soft Green
    "nlp_sentence_transformer_question_tags": "#3F7D20"  # Deep Green
}

# Ensure the categorical order to display "top_users" first in the legend
df_results["type"] = pd.Categorical(df_results["type"], categories=list(custom_palette.keys()), ordered=True)

# Create the bar plot with the updated color palette
plt.figure(figsize=(20, 8))  # Keeping consistent styling
ax = sns.barplot(
    data=df_results,
    x="top",
    y="acc",
    hue="type",
    palette=custom_palette  # Apply the new custom palette
)

# Modify labels and title
plt.title("Accuracy Across Different Top-N Recommendations", fontsize=24, fontweight="bold", pad=20)
plt.xlabel("Top-N Recommendations", fontsize=20, labelpad=10)
plt.ylabel("Hit Accuracy", fontsize=20, labelpad=10)

# Update legend labels to ensure correct naming order
legend_labels = {
    "top_users": "Baseline", 
    "collaborative": "S1 - Collaborative Filtering", 
    "content": "S2 - Content-Based Filtering", 
    "nlp_sentence_transformer_user_profile": "S3 - User Profile (NLP)",
    "nlp_sentence_transformer_question": "S4 - Question (NLP)",
    "nlp_sentence_transformer_question_tags": "S5 - Question + Tags (NLP)"
}
handles, labels = ax.get_legend_handles_labels()
new_labels = [legend_labels[label] if label in legend_labels else label for label in labels]

# Reorder legend to ensure "Top Users" appears first
sorted_handles_labels = sorted(zip(handles, new_labels), key=lambda x: list(legend_labels.values()).index(x[1]))
handles, new_labels = zip(*sorted_handles_labels)

plt.legend(handles, new_labels, title="Recommendation Type", fontsize=18, title_fontsize=20, loc="upper left", bbox_to_anchor=(1, 1))

# Improve spacing
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Adjust layout to make space for the legend

# Remove top and right borders
sns.despine()

# Show the plot
plt.show()

```
